```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = dataset.drop("Health Issue", axis=1)
y = dataset['Health Issue']
```

```python
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y, test_size=0.3, random_state=20)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd

# Assume X and y are your full dataset and labels (replace them with your actual dataset)
# Example dataset initialization
# X = pd.DataFrame({ ... }) # Your dataset goes here
# y = pd.Series(...)

# Concatenate train and test datasets to fit the LabelEncoder on all the data
X_full = pd.concat([X_train, X_test])

# Convert non-numeric (categorical) columns to numeric using LabelEncoder
le = LabelEncoder()

# Apply LabelEncoder to the full dataset before splitting
X_full_encoded = X_full.apply(le.fit_transform)

# Now split back into training and testing sets
X_train_encoded = X_full_encoded.loc[X_train.index]
X_test_encoded = X_full_encoded.loc[X_test.index]

# Create a dictionary to store models
models = {
    "SVC": SVC(kernel='linear'),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "KNeighbors": KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB": MultinomialNB()
}

# Train and test each model
for model_name, model in models.items():
    # Train model
    model.fit(X_train_encoded, y_train)

    # Test model
    predictions = model.predict(X_test_encoded)

    # Calculate confusion matrix and accuracy
    cm = confusion_matrix(y_test, predictions)
    accuracy = accuracy_score(y_test, predictions)

    print(f"{model_name} accuracy: {accuracy}")
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', '))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Fit the LabelEncoder on the combined dataset (train + test)
le = LabelEncoder()

# Concatenate train and test sets temporarily for encoding
X_combined = pd.concat([X_train, X_test])

# Apply Label Encoding to all columns with string data in the combined dataset
X_combined_encoded = X_combined.apply(lambda col: le.fit_transform(col) if col.dtype == 'object' else col)

# Split the combined dataset back into train and test sets
X_train_encoded = X_combined_encoded.iloc[:len(X_train), :]
X_test_encoded = X_combined_encoded.iloc[len(X_train):, :]

# Corrected 'kernel' parameter
svc = SVC(kernel='linear')

# Fit the model using encoded data
svc.fit(X_train_encoded, y_train)

# Make predictions
ypred = svc.predict(X_test_encoded)

# Calculate accuracy
accuracy = accuracy_score(y_test, ypred)
print("SVC Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt

def plot_prakruti(vata_score, pitta_score, kapha_score):
    labels = ['Vata', 'Pitta', 'Kapha']
    scores = [vata_score, pitta_score, kapha_score]

    plt.bar(labels, scores, color=['blue', 'red', 'green'])
    plt.xlabel('Dosha Types')
    plt.ylabel('Scores')
    plt.title('Prakruti Assessment Result')
    plt.show()
```

```python
# Call the function after getting scores
vata_score, pitta_score, kapha_score = 3, 5, 2  # Example scores, replace with actual results
plot_prakruti(vata_score, pitta_score, kapha_score)
```
